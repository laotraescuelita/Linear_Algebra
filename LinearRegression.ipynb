```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# importar de sklearn el modulo para crear de manera artificial matrices para clasificaci√≥n.
from sklearn.datasets import make_regression
# Generar la matriz con variables numericas y el vector a predecir
A, b, coefficients = make_regression(
n_samples = 100,
n_features = 100,
n_informative = 3,
n_targets = 1,
noise = 0.2,
coef = True,
random_state = 1)
```

```python
# Resolvamos Ax=b utilizando Singular Value Decomposition
U, S, VT = np.linalg.svd(A,full_matrices=0)
x = VT.T @ np.linalg.inv(np.diag(S)) @ U.T @ b
```

```python
#Mostrar la forma que tienen las variables 
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].plot(b, label='Datos sinteticos') # Valores "reales".
axes[0].set_xlabel('Eje de las x')
axes[0].set_ylabel('Eje de las y')
axes[0].legend()

axes[1].plot(A@x, '-o',  label='Singular value decomposition') #Valores predecidos.
axes[1].set_xlabel('Eje de las x')
axes[1].set_ylabel('Eje de la y')
axes[1].legend()
plt.show()

#Las graficas se parecen demasiado porque los valores sinteticos estan muy limpios, no hay demsiado ruido.
```

```python
#Mostrar el resultado pero con datos aleatorios de la variable a predecir
sort_ind = np.argsort(b)
b = b[sort_ind] # sorted values
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].plot(b,  label='Datos sinteticos') # Relaci√≥n original de los datos
axes[0].set_xlabel('Eje de las x')
axes[0].legend()

axes[1].plot(A[sort_ind,:]@x, '-x', label='Singular value decomposition')
axes[1].set_xlabel('Eje de las x')
axes[1].legend()

plt.show()
```
