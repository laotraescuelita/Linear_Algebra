```python
import numpy as np 
import matplotlib.pyplot as plt 
from matplotlib.image import imread
```

```python
def svd_1( img ):
    
    img_gray = np.mean( img , -1)
    
    print( "El tamaño de la imagen original es", img.shape)
    plt.imshow( img )
    plt.title( "Imagen original ")
    plt.show()

    print( "El tamaño de la imagen en escalas de grises", img_gray.shape)
    plt.imshow( img_gray )
    plt.title( "Imagen en escala de grises ")
    plt.show()
    
    #Método para comrprimir la imagen.
    U,S,V = np.linalg.svd( img_gray, full_matrices = False )
    S_D = np.diag( S )

    #vamos a mostrar las iamgenes después de la transformaión.
    fig, axe = plt.subplots(1,3,figsize=(20,4))    
    
    img_aprox = U[:,:5] @ S_D[0:5,:5] @ V[:5,:]       
    axe[0].imshow(img_aprox,cmap="gray")
    axe[0].axis("off")
    axe[0].set_title("Hasta 5 columnas")
    
    img_aprox = U[:,:20] @ S_D[0:20,:20] @ V[:20,:]
    axe[1].imshow(img_aprox,cmap="gray")
    axe[1].axis("off")
    axe[1].set_title("Hasta 20 columnas")
    
    img_aprox = U[:,:20] @ S_D[0:20,:20] @ V[:20,:]
    axe[2].imshow(img_aprox, cmap="gray")
    axe[2].axis("off")
    axe[2].set_title("Hasta 100 columnas")
    
    
        
```

```python
img = imread( "img.jpg")
svd_1( img )
```
