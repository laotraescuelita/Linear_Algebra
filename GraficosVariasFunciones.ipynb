```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Funciones lineales

#En el caso de que las funciones son constantes. 
#Ejemplo  x = 3 
y = np.linspace(-5,5,5)
x = [3]*5
plt.plot(x,y,c="red")
plt.title("X constante")
plt.xlabel("Eje de las x´s")
plt.ylabel("Eje de las y´s")
plt.grid(color="green")
plt.xlim(-5,5)
plt.ylim(-5,5)
```

```python
#### Utilizando la forma  y = mx + b 
x = np.linspace(-5,5,5)
def formapendiente( x, pendiente, intercepto ):
    return pendiente*x + intercepto

fig, axe = plt.subplots(1,5,figsize=(20,4))
intercepto = [-4,-1,0,1,4]
for i in range(len(intercepto) ):
    #Vamos a mirar que sucede cuando el intercepto se mueve y la pendiente es la misma
    y = formapendiente( x,2,intercepto[i] )
    axe[i].plot(x,y,c="red")
    axe[i].set_title("Intercepto {} ".format( intercepto[i] ))
    axe[i].set_xlabel("Eje de las x´s")
    axe[i].set_ylabel("Eje de las y´s")
    axe[i].grid(color="green")
    axe[i].set_xlim(-5,5)
    axe[i].set_ylim(-5,5)

fig, axe = plt.subplots(1,5,figsize=(20,4))
pendiente = [-5,-1,0,1,5]
for i in range(len(intercepto) ):
    #Vamos a mirar que sucede cuando el intercepto se mueve
    y = formapendiente( x,pendiente[i],0 )
    axe[i].plot(x,y,c="red")
    axe[i].set_title("Pendiente {} ".format( pendiente[i] ))
    axe[i].set_xlabel("Eje de las x´s")
    axe[i].set_ylabel("Eje de las y´s")
    axe[i].grid(color="green")
    axe[i].set_xlim(-5,5)
    axe[i].set_ylim(-5,5)
```

```python
#Miremos las formas de las funciones cuadraticas
x = np.linspace(-5,5,50)

fig1, axe = plt.subplots(1,4,figsize=(20,4))

def formacuadratica( x ):
    return x**2

y = formacuadratica( x )
axe[0].plot(x,y,c="red")
axe[0].set_title("Forma {} ".format( "x**2" ))
axe[0].set_xlabel("Eje de las x´s")
axe[0].set_ylabel("Eje de las y´s")
axe[0].grid(color="green")
axe[0].set_xlim(-5,5)
axe[0].set_ylim(-5,5)

def formacuadratica( x ):
    return -(x**2)

y = formacuadratica( x )
axe[1].plot(x,y,c="red")
axe[1].set_title("Forma {} ".format( "-x**2" ))
axe[1].set_xlabel("Eje de las x´s")
axe[1].set_ylabel("Eje de las y´s")
axe[1].grid(color="green")
axe[1].set_xlim(-5,5)
axe[1].set_ylim(-5,5)

def formacuadratica( x ):
    return (x**2) + 3

y = formacuadratica( x )
axe[2].plot(x,y,c="red")
axe[2].set_title("Forma {} ".format( "x**2 + 3" ))
axe[2].set_xlabel("Eje de las x´s")
axe[2].set_ylabel("Eje de las y´s")
axe[2].grid(color="green")
axe[2].set_xlim(-5,5)
axe[2].set_ylim(-5,5)

def formacuadratica( x ):
    return x**2 - 3

y = formacuadratica( x )
axe[3].plot(x,y,c="red")
axe[3].set_title("Forma {} ".format( "x**2-3" ))
axe[3].set_xlabel("Eje de las x´s")
axe[3].set_ylabel("Eje de las y´s")
axe[3].grid(color="green")
axe[3].set_xlim(-5,5)
axe[3].set_ylim(-5,5)
```

```python
#Como podemos entender las graficas de la función valor absoluto 
x = np.linspace(-5,5,50)

fig = plt.figure(figsize=(16,4))

def valorabsoluto( x ):
    return np.absolute( x )
y = valorabsoluto( x )

ax1 = fig.add_subplot(1, 3, 1)
ax1.plot(x,y,c="red")
ax1.set_title("Valor absoluto {} ".format( "|x|" ))
ax1.set_xlabel("Eje de las x´s")
ax1.set_ylabel("Eje de las y´s")
ax1.grid(color="green")
ax1.set_xlim(-5,5)
ax1.set_ylim(-5,5)

def valorabsoluto( x ):
    return -np.absolute( x )
y = valorabsoluto( x )

ax2 = fig.add_subplot(1, 3, 2)
ax2.plot(x,y,c="red")
ax2.set_title("Valor absoluto {} ".format( "-|x|" ))
ax2.set_xlabel("Eje de las x´s")
ax2.set_ylabel("Eje de las y´s")
ax2.grid(color="green")
ax2.set_xlim(-5,5)
ax2.set_ylim(-5,5)

#Las grafica se pueden mover del origen a otras coordinadas asi como modificar al escala con operaciones aritmeticas. 
#por ejemplo y = 5 - 3 * |x|
def valorabsoluto( x ):
    return -3*np.absolute( x ) + 5
y = valorabsoluto( x )

ax3 = fig.add_subplot(1, 3, 3)
ax3.plot(x,y,c="red")
ax3.set_title("Valor absoluto {} ".format( "-3*|x|+5" ))
ax3.set_xlabel("Eje de las x´s")
ax3.set_ylabel("Eje de las y´s")
ax3.grid(color="green")
ax3.set_xlim(-5,5)
ax3.set_ylim(-5,5)
```

```python
#Miremos las formas de las funciones radicales 
x = np.linspace(-5,5,50)

fig1, axe = plt.subplots(1,4,figsize=(20,4))

def formaradical( x ):
    return np.sqrt( x )

y = formaradical( x )
axe[0].plot(x,y,c="red")
axe[0].set_title("Forma {} ".format( "raiz( x )" ))
axe[0].set_xlabel("Eje de las x´s")
axe[0].set_ylabel("Eje de las y´s")
axe[0].grid(color="green")
axe[0].set_xlim(-5,5)
axe[0].set_ylim(-5,5)

def formaradical( x ):
    return - np.sqrt( x )

y = formaradical( x )
axe[1].plot(x,y,c="red")
axe[1].set_title("Forma {} ".format( " - raiz(x) " ))
axe[1].set_xlabel("Eje de las x´s")
axe[1].set_ylabel("Eje de las y´s")
axe[1].grid(color="green")
axe[1].set_xlim(-5,5)
axe[1].set_ylim(-5,5)

def formaradical( x ):
    return  np.sqrt( -x )

y = formaradical( x )
axe[2].plot(x,y,c="red")
axe[2].set_title("Forma {} ".format( " raiz( - x ) " ))
axe[2].set_xlabel("Eje de las x´s")
axe[2].set_ylabel("Eje de las y´s")
axe[2].grid(color="green")
axe[2].set_xlim(-5,5)
axe[2].set_ylim(-5,5)

def formaradical( x ):
    return  - np.sqrt( -x )

y = formaradical( x )
axe[3].plot(x,y,c="red")
axe[3].set_title("Forma {} ".format( " - raiz( - x ) " ))
axe[3].set_xlabel("Eje de las x´s")
axe[3].set_ylabel("Eje de las y´s")
axe[3].grid(color="green")
axe[3].set_xlim(-5,5)
axe[3].set_ylim(-5,5)
```

```python
#Miremos las formas de las funciones racionales 
x = np.linspace(-5,5,50)

fig1, axe = plt.subplots(1,4,figsize=(20,4))

def formaracional( x ):
    return  1/x 

y = formaracional( x )
axe[0].plot(x,y,c="red")
axe[0].set_title("Forma {} ".format( "( 1/x )" ))
axe[0].set_xlabel("Eje de las x´s")
axe[0].set_ylabel("Eje de las y´s")
axe[0].grid(color="green")
axe[0].set_xlim(-5,5)
axe[0].set_ylim(-5,5)

def formaracional( x ):
    return -( 1/x )

y = formaracional( x )
axe[1].plot(x,y,c="red")
axe[1].set_title("Forma {} ".format( " - ( 1/x ) " ))
axe[1].set_xlabel("Eje de las x´s")
axe[1].set_ylabel("Eje de las y´s")
axe[1].grid(color="green")
axe[1].set_xlim(-5,5)
axe[1].set_ylim(-5,5)

def formaracional( x ):
    return  ( 1/-x )

y = formaracional( x )
axe[2].plot(x,y,c="red")
axe[2].set_title("Forma {} ".format( " ( 1/-x ) " ))
axe[2].set_xlabel("Eje de las x´s")
axe[2].set_ylabel("Eje de las y´s")
axe[2].grid(color="green")
axe[2].set_xlim(-5,5)
axe[2].set_ylim(-5,5)

def formaracional( x ):
    return  - ( 1/-x )

y = formaracional( x )
axe[3].plot(x,y,c="red")
axe[3].set_title("Forma {} ".format( " - ( 1/-x ) " ))
axe[3].set_xlabel("Eje de las x´s")
axe[3].set_ylabel("Eje de las y´s")
axe[3].grid(color="green")
axe[3].set_xlim(-5,5)
axe[3].set_ylim(-5,5)
```
